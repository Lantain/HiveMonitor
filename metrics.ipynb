```python
import matplotlib.pyplot as plt

filtered_data = ds[ds['hive number'] == 3]

plt.figure(figsize=(60, 6))  # Adjust the width as per your requirement
plt.plot(filtered_data['date'], filtered_data['hive temp'])
plt.xlabel('Date')
plt.ylabel('Hive Temperature')
plt.title('Hive Temperature over Time')
plt.xticks(rotation=45)

plt.show()
```
