```python

import pandas as pd
import dataset
import os

SRC_DIR = f'dataset/transformed_timestamps/interval_15m'
OUT_DIR = f'dataset/transformed_populated/{SRC_DIR.split("/")[-1]}'
os.makedirs(OUT_DIR, exist_ok=True)
files = os.scandir(SRC_DIR)
```

```python
for file in files:
    print(f"Processing {file.name}")
    
    df = dataset.read_dataset_file(file.path)
    df.infer_objects(copy=False)                
    df.bfill()
    print(f"  Smoothing")

    df['weight_kg_smoothed'] = dataset.smooth_col(df['weight_kg'])
    

    
    df['temp_mid'] = df[['t_i_1', 't_i_2', 't_i_3', 't_i_4', 't_i_5']].median(axis=1, skipna=True)
    
    df['t_i_1'] = dataset.smooth_col(df['t_i_1'])
    df['t_i_2'] = dataset.smooth_col(df['t_i_2'])
    df['t_i_3'] = dataset.smooth_col(df['t_i_3'])
    df['t_i_4'] = dataset.smooth_col(df['t_i_4'])
    df['t_i_5'] = dataset.smooth_col(df['t_i_5'])

    df['t'] = dataset.smooth_col(df['t'])
    df['h'] = dataset.smooth_col(df['h'])
    df['p'] = dataset.smooth_col(df['p'])
    print(f"  Populating")
    df['weight_kg_3_pct'] = df['weight_kg_smoothed'].pct_change(periods=3)
    df['weight_kg_5_pct'] = df['weight_kg_smoothed'].pct_change(periods=5)
    df['weight_kg_10_pct'] = df['weight_kg_smoothed'].pct_change(periods=10)
    
    dataset.populate_delta(df)
    dataset.populate_temp_delta(df)
    dataset.populate_humidity_delta(df)
    
    df.to_csv(f'{OUT_DIR}/{file.name}', index=False)
```
