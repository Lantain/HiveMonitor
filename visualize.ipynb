```python
import pandas as pd
from pandas import DataFrame
import math
import dataset
import main
import events
import os
import numpy as np

SRC_DIR =  'dataset/b_parsed_20m/parsed/feeding'

columns = [
    # 'temp_mid',
    # 'h', 
    'weight_kg', 
    'weight_perc_l3',
    # 't_perc_l10',
    # 'h_perc_l10',
]
```

```python
files = os.listdir(SRC_DIR)
index = 10

print(f'{index}/{len(files)}')
file = files[index]
df = pd.read_csv(f'{SRC_DIR}/{file}')

try:
    swarming_indexes = df[df['swarming'] == 1].index
    queencell_indexes = df[df['queencell'] == 1].index
    feeding_indexes = df[df['feeding'] == 1].index
    honey_indexes = df[df['honey'] == 1].index
    treatment_indexes = df[df['treatment'] == 1].index
    died_indexes = df[df['died'] == 1].index
except:
    swarming_indexes = []
    queencell_indexes = []
    feeding_indexes = []
    honey_indexes = []
    treatment_indexes = []
    died_indexes = []

print (f'SW {len(swarming_indexes)},    QC {len(queencell_indexes)},    F {len(feeding_indexes)},   H {len(honey_indexes)},   T   {len(treatment_indexes)},    D {len(died_indexes)}')
for col in columns:
    title = os.path.basename(file) + ' == ' + col
    main.show_graph(df[col], title, swarming_indexes, queencell_indexes, feeding_indexes, honey_indexes, treatment_indexes, died_indexes)
```
