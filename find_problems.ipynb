```python
import pandas as pd
from pandas import DataFrame
import math
import os
from tensorflow import keras
from keras import layers, models, optimizers
import numpy as np
from sklearn.utils import shuffle
import dataset
import events
import main

SEGMENT_SIZE = 30
COLUMN = 'weight_perc_l3'

SRC_DIR = 'dataset/bobo/swarming_'+ '20m'
x, y = dataset.dir_to_split(SRC_DIR, COLUMN, SEGMENT_SIZE)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sample_sequences(x, y, num_samples=5, figsize=(15, 10)):
    """
    Plot sample sequences from the dataset.
    
    Args:
        x: Input sequences of shape (n_samples, sequence_length, features)
        y: Target values of shape (n_samples, 1)
        num_samples: Number of random samples to plot
        figsize: Size of the figure
    """
    # Randomly select samples
    indices = np.random.choice(x.shape[0], num_samples, replace=False)
    
    # Create subplot grid
    fig, axes = plt.subplots(num_samples, 1, figsize=figsize)
    fig.suptitle('Sample Sequences with Target Values', fontsize=16)
    
    for i, idx in enumerate(indices):
        # Plot sequence
        axes[i].plot(range(x.shape[1]), x[idx, :, 0], 'b-', label='Input Sequence')
        
        # Plot target value as a horizontal line
        axes[i].axhline(y=y[idx], color='r', linestyle='--', 
                       label=f'Target Value: {y[idx][0]:.2f}')
        
        axes[i].set_title(f'Sample {idx}')
        axes[i].set_xlabel('Time Step')
        axes[i].set_ylabel('Value')
        axes[i].legend()
        axes[i].grid(True)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_sample_sequences(x, y)
```
