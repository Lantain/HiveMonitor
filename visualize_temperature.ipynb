```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV data
df = pd.read_csv('dataset/bob_publication_data/years/2021/preprocessed/2021_d/26.csv')

# Convert the 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set the style
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Create the line plot
sns.lineplot(data=df, 
            x='time', 
            y='t_i_1',
            linewidth=2,
            color='#6F1D1B')
sns.lineplot(data=df, 
            x='time', 
            y='t_i_2',
            linewidth=2,
            color='#BB9457')
sns.lineplot(data=df, 
            x='time', 
            y='t_i_3',
            linewidth=2,
            color='#432818')
sns.lineplot(data=df, 
            x='time', 
            y='t_i_4',
            linewidth=2,
            color='#99582A')
sns.lineplot(data=df, 
            x='time', 
            y='t_i_5',
            linewidth=2,
            color='#FFE6A7')
sns.lineplot(data=df, 
            x='time', 
            y='t_o',
            linewidth=2,
            color='#4F6367')

# Customize the plot
plt.title('Temperature (t_i_1) Changes Over Time', pad=15, size=14)
plt.xlabel('Date', size=12)
plt.ylabel('Temperature (°C)', size=12)
plt.legend(['t_i_1', 't_i_2', 't_i_3', 't_i_4', 't_i_5', 't_o'], loc='upper left')
# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add margins to prevent label cutoff
plt.margins(x=0.02)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```
